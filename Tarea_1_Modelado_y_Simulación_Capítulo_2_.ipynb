##**2.1.3**

a) Encontrar la aceleración del flujo $\ddot{x}$ en función de $x$.

**Solución:**

Podemos encontrar la aceleración del flujo x derivando primero una ecuación que contiene $\ddot{x}$ tomando la derivada temporal de la ecuación diferencial.

\begin{equation}
\frac{d}{dt}\dot{x}=\frac{d}{dt}sin\left(x\right)
\longrightarrow\ \ddot{x}\ =\ cos(x)\ \dot{x}
\end{equation}

Podemos obtener $\ddot{x}$ únicamente en función de x conectando nuestra ecuación anterior par $\dot{x}$

\begin{equation}
\ddot{x}=\ cos(x)\ sin(x)
\end{equation}


b) Encontrar los puntos donde el flujo tiene la máxima aceleración positiva.

Podemos encontrar qué valores de x dan a la aceleración $\ddot{x}$ valores positivos máximos utilizando la identidad trigonométrica.

\begin{equation}
\frac{1}{2}\ sin(2x)\ =\ sin(x)\ cos(x)
\end{equation}

que se puede utilizar para reescribir ̈x como

\begin{equation}
\ddot{x}=\frac{1}{2}\ sin(2x)
\end{equation}

que tiene máximos cuando 

\begin{equation}
x=\left(z+\frac{1}{4}\right)\pi,\ \ z\ \in\ \mathbb Z
\end{equation}



```python
def step_euler(y,h,f):
  next_y=y+h*f(y)
  return next_y

def euler (h,f,y_0,start_t,end_t):
  N= int((end_t - start_t)/h)
  t_list=np.linspace(start_t,end_t,N+1)

  y_list=np.zeros(N+1)

  y_list[0]=y_0

  for i in range(0,N):
    y_list[i+1]=step_euler(y_list[i],h,f)

  return t_list,y_list

def Ejemplo2_1(x):
  dx=np.sin(x)
  return dx
y_0=0
h=0.1
t_0=0
t_N=10
for i in np.arange(0,6,0.2):
  t_list,y_list=euler(h,Ejemplo2_1,i, t_0,t_N)
  plt.plot(t_list,y_list, label='Euler')

#plt.plot(t_list, np.power(2,t_list),label='Euler')
#plt.legend()
print(np.abs(y_list[-1]-np.power(2,t_list[-1])))
plt.grid()
plt.axis([0,10,0,6])
```

## **2.3.3** 
(Crecimiento tumoral) El crecimiento de tumores cancerosos puede ser modelado por la ley de Gompertz $\dot{N}=-aNln(bN)$, donde $N ( t )$ es proporcional al número de células en el tumor, y $a, b > 0$ son parámetros.\
 a) Interpretar a y b biológicamente.\
 **Solucion**\
En un primer momento se puede interpretar $a$ como un parámetro que información acerca de la rapidez con la que crece el tumor. Por otro lado, tenemos que $\frac{1}{b} $ indaga sobre del tamaño del tumor.\
 b) Esboce el campo vectorial y luego grafique $N ( t )$ para varios valores iniciales. Las predicciones de este modelo simple concuerdan sorprendentemente bien con los datos sobre el crecimiento tumoral, siempre y cuando N no sea demasiado pequeño; ver Aroesty et al. (1973) y Newton (1980) para ejemplos.

 **Solución Analitica**
\begin{equation}
\dot{N}=-aNln(bN)\\
\frac{dN}{dt}=-aNln(bN)\\
\int\frac{dN}{-aN\ln(bN)}=\int{dt}\\
t=\int\frac{dN}{-aN\ln(bN)}
\end{equation}

Se aplica la sustitución
\begin{equation}
u=\ln(bN)\rightarrow\frac{du}{dN}=\frac{1}{N}\rightarrow dN=Ndu\\
t=-\frac{1}{a}\int \frac{1}{u}du\\
t=-\frac{\ln{u}}{a}
\end{equation}

Se deshace la sustitución

\begin{equation}
