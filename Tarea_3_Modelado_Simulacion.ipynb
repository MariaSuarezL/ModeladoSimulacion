## 5.1.3 $\dot{x}=-y, \dot{y}=−x$ 

\begin{equation}
\left[\begin{matrix}\dot{x}\\\dot{y}\\\end{matrix}\right]=\left[\begin{matrix}0&-1\\-1&0\\\end{matrix}\right] \left[\begin{matrix}x\\y\\\end{matrix}\right]
\end{equation}

**Polinomio caracteristico**

\begin{equation}
det\left[\begin{matrix}-\lambda&-1\\-1&-\lambda\\\end{matrix}\right] =0
\end{equation}

\begin{equation}
\lambda^2-1=0
\end{equation}

**Valores propios**
\begin{equation}
\lambda_{1}=1,\lambda_{2}=-1
\end{equation}

**Vectores propios**

\begin{equation}
(A-\lambda_{1}I)\vec{v_1}=\vec{0}\\
\left[\begin{matrix}-\lambda_{1}&-1\\-1&-\lambda_{1}\\\end{matrix}\right]\vec{v_1}=\vec{0}\\
\left[\begin{matrix}-1&-1\\-1&-1\\\end{matrix}\right]\vec{v_1}=\vec{0}\\
\vec{v_1}=\left[\begin{matrix}-1\\1\\\end{matrix}\right]
\end{equation}


\begin{equation}
(A-\lambda_{2}I)\vec{v_2}=\vec{0}\\
\left[\begin{matrix}-\lambda_{2}&-1\\-1&-\lambda_{2}\\\end{matrix}\right]\vec{v_2}=\vec{0}\\
\left[\begin{matrix}1&-1\\-1&1\\\end{matrix}\right]\vec{v_2}=\vec{0}\\
\vec{v_2}=\left[\begin{matrix}1\\1\\\end{matrix}\right]
\end{equation}

**Solución general del sistema**

\begin{equation}
\left[\begin{matrix}x\\y\\\end{matrix}\right]=c_{1}e^{\lambda_{1}t}\vec{v_1}+c_{2}e^{\lambda_{2}t}\vec{v_2}=c_{1}e^{t}\left[\begin{matrix}-1\\1\\\end{matrix}\right]+c_{2}e^{-t}\left[\begin{matrix}1\\1\\\end{matrix}\right]
\end{equation}

**Clasifica el punto fijo en el origen: Saddle point**

\begin{equation}
\tau=0\\
\Delta=-1<0
\end{equation}

\begin{equation}
\tau^2-4\Delta=0^2-4(-1)=0+4>0
\end{equation}



```python
def fun (t,xy,a,b,c,d):
  x,y=xy
  return [a*x+b*y,c*x+d*y]
  
a,b,c,d=0,-1,-1,0

yi=-10; yf=10
xi=-10; xf=10
ti=0; tf=10; dt=1000
tspam=[ti, tf]
t=np.linspace(ti,tf,dt)
for x0 in np.arange(xi,xf,1):
  for y0 in np.arange(yi,yf,1):
    sol=solve_ivp(fun,tspam,[x0,y0],args=(a,b,c,d),t_eval=t)
    plt.plot(sol.y[0], sol.y[1])

plt.axis([-10, 10, -10, 10])
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

# Creating data set
w = 10
Y, X = np.mgrid[-w:w:100j, -w:w:100j]
U = a*X+b*Y
V = c*X+d*Y
speed = np.sqrt(U**2 + V**2)
 
# Creating plot
plt.streamplot(X, Y, U, V, density = 1,color='gray')
 
# show plot
plt.show()
```

##5.2.1 $\dot{x}=4x-y,\dot{y}=2x+y$


Considere el sistema $\dot{x}=4x-y,\dot{y}=2x+y$

**a)** Escriba el sistema como $\dot{x}=Ax$ . Demostrar que el polinomio característico es
$\lambda^2-5\lambda+6$, y encuentre los valores y vectores propios de A.

**Sistema en forma matricial**

\begin{equation}
\left[\begin{matrix}\dot{x}\\\dot{y}\\\end{matrix}\right]=\left[\begin{matrix}4&-1\\2&1\\\end{matrix}\right] \left[\begin{matrix}x\\y\\\end{matrix}\right]
\end{equation}

**Polinomio caracteristico**

\begin{equation}
det\left[\begin{matrix}4-\lambda&-1\\2&1-\lambda\\\end{matrix}\right] =0
\end{equation}

\begin{equation}
(4-\lambda)(1-\lambda)-(2)(-1)=0\\
\lambda^2-5\lambda+6=0
\end{equation}

\begin{equation}
\tau=5\\
\Delta=6
\end{equation}

**Valores propios**
\begin{equation}
\lambda^2-5\lambda+6=0\\
(\lambda-2)(\lambda-3)=0\\
\lambda_{1}=2,\lambda_{2}=3
\end{equation}

**Vectores propios**

\begin{equation}
(A-\lambda_{1}I)\vec{v_1}=\vec{0}\\
\left[\begin{matrix}4-\lambda_{1}&-1\\2&1-\lambda_{1}\\\end{matrix}\right]\vec{v_1}=\vec{0}\\
\left[\begin{matrix}2&-1\\2&-1\\\end{matrix}\right]\vec{v_1}=\vec{0}\\
\vec{v_1}=\left[\begin{matrix}1\\2\\\end{matrix}\right]
\end{equation}


\begin{equation}
(A-\lambda_{2}I)\vec{v_2}=\vec{0}\\
\left[\begin{matrix}4-\lambda_{2}&-1\\2&1-\lambda_{2}\\\end{matrix}\right]\vec{v_2}=\vec{0}\\
\left[\begin{matrix}1&-1\\2&-2\\\end{matrix}\right]\vec{v_2}=\vec{0}\\
\vec{v_2}=\left[\begin{matrix}1\\1\\\end{matrix}\right]
\end{equation}

**b)** Hallar la solución general del sistema.

\begin{equation}
\left[\begin{matrix}x\\y\\\end{matrix}\right]=c_{1}e^{\lambda_{1}t}\vec{v_1}+c_{2}e^{\lambda_{2}t}\vec{v_2}=c_{1}e^{2t}\left[\begin{matrix}1\\2\\\end{matrix}\right]+c_{2}e^{3t}\left[\begin{matrix}1\\1\\\end{matrix}\right]
\end{equation}

**c)** Clasifica el punto fijo en el origen.

\begin{equation}
\tau=5>0\\
\Delta=6>0
\end{equation}

\begin{equation}
\tau^2-4\Delta=5^2-4(6)=25-24=1>0
\end{equation}

Es un nodo inestable 

**d)** Resolver el sistema sujeto a la condición inicial $( x0,y0)=(3, 4)$.

\begin{equation}
\left[\begin{matrix}1&1\\2&1\\\end{matrix}\right] \left[\begin{matrix}c_1\\c_2\\\end{matrix}\right]=\left[\begin{matrix}3\\4\\\end{matrix}\right]\\
c_1=1,c_2=2
\end{equation}

El sistema sujeto a estos valores iniciales tendrian como solucion:

\begin{equation}
\left[\begin{matrix}x\\y\\\end{matrix}\right]=e^{2t}\left[\begin{matrix}1\\2\\\end{matrix}\right]+2e^{3t}\left[\begin{matrix}1\\1\\\end{matrix}\right]

##5.2.2  $\dot{x}=x−y, \dot{y}=x+y$
(Valores propios complejos) Este ejercicio lo lleva a través de la solución de un sistema lineal donde los valores propios son complejos. El sistema es $\dot{x}=x−y, \dot{y}=x+y$

**a)** Encuentre $A$ y demuestre que tiene valores propios $\lambda_{1}=1+i, \lambda_{1}=1-i $, con vectores propios $\vec{v_1}( i , 1),\vec{v_2}( -i , 1)$. (Tenga en cuenta que los valores propios son conjugados complejos,y también son los vectores propios; este es siempre el caso para el real A con valores propios complejos.)


**Sistema en forma matricial**

\begin{equation}
\left[\begin{matrix}\dot{x}\\\dot{y}\\\end{matrix}\right]=\left[\begin{matrix}1&-1\\1&1\\\end{matrix}\right] \left[\begin{matrix}x\\y\\\end{matrix}\right]
\end{equation}

**Polinomio caracteristico**

\begin{equation}
det\left[\begin{matrix}1-\lambda&-1\\1&1-\lambda\\\end{matrix}\right] =0
\end{equation}

\begin{equation}
(1-\lambda)(1-\lambda)-(1)(-1)=0\\
\lambda^2-2\lambda+2=0
\end{equation}

\begin{equation}
\tau=2\\
\Delta=2
\end{equation}

**Valores propios**
\begin{equation}
\lambda^2-2\lambda+2=0\\
\lambda_{1}=\frac{2+\sqrt{2^2-4(2)}}{2}=\frac{2+2i}{2}=1+i\\
\lambda_{2}=\frac{2-\sqrt{2^2-4(2)}}{2}=\frac{2-2i}{2}=1-i
\end{equation}

**Vectores propios**

\begin{equation}
(A-\lambda_{1}I)\vec{v_1}=\vec{0}\\
\left[\begin{matrix}1-\lambda_{1}&-1\\1&1-\lambda_{1}\\\end{matrix}\right]\vec{v_1}=\vec{0}\\
\left[\begin{matrix}-i&-1\\1&-i\\\end{matrix}\right]\vec{v_1}=\vec{0}\\
\vec{v_1}=\left[\begin{matrix}i\\1\\\end{matrix}\right]
\end{equation}


\begin{equation}
(A-\lambda_{2}I)\vec{v_2}=\vec{0}\\
\left[\begin{matrix}1-\lambda_{2}&-1\\1&1-\lambda_{2}\\\end{matrix}\right]\vec{v_2}=\vec{0}\\
\left[\begin{matrix}i&-1\\1&i\\\end{matrix}\right]\vec{v_2}=\vec{0}\\
\vec{v_2}=\left[\begin{matrix}-i\\1\\\end{matrix}\right]
\end{equation}


**b)** La solución general es $x(t)=c_{1}e^{\lambda_{1}t}\vec{v_1}+c_{2}e^{\lambda_{2}t}\vec{v_2}$.Así que en cierto sentido hemos terminado.Pero esta forma de escribir $x(t)$ involucra coeficientes complejos y parece poco familiar.
Exprese $x( t )$ puramente en términos de funciones de valor real. (Sugerencia: utilice $e^{iwt}=cos(wt) + i*sin(wt) $ para reescribir $x( t )$ en términos de senos y cosenos, y luego separar los términos
que tienen un prefactor de i de los que no).

\begin{equation}
x(t)=c_{1}e^{\lambda_{1}t}\vec{v_1}+c_{2}e^{\lambda_{2}t}\vec{v_2}\\
x(t)=c_{1}e^{(1+i)t}\left[\begin{matrix}i\\1\\\end{matrix}\right]+c_{2}e^{(1-i)t}\left[\begin{matrix}-i\\1\\\end{matrix}\right]\\
x(t)=c_{1}e^{t}e^{it}\left[\begin{matrix}i\\1\\\end{matrix}\right]+c_{2}e^{t}e^{-it}\left[\begin{matrix}-i\\1\\\end{matrix}\right]\\
x(t)=c_{1}e^{t}(cos(t) + i*sin(t))\left[\begin{matrix}i\\1\\\end{matrix}\right]+c_{2}e^{t}(cos(-t) + i*sin(-t))\left[\begin{matrix}-i\\1\\\end{matrix}\right]\\
x(t)=\left[\begin{matrix}ic_{1}e^{t}(cos(t) + i*sin(t))\\c_{1}e^{t}(cos(t) + i*sin(t))\\\end{matrix}\right]+\left[\begin{matrix}-ic_{2}e^{t}(cos(t) - i*sin(t))\\c_{2}e^{t}(cos(t) - i*sin(t))\\\end{matrix}\right]\\
x(t)=e^{t}\left[\begin{matrix}icos(t)(c_1-c_2)-sin(t)(c_1+c_2)\\cos(t)(c_1+c_2)+isin(t)(c_1-c_2)\\\end{matrix}\right]
\end{equation}

**Clasifica el punto fijo en el origen=Espiral inestable**

\begin{equation}
\tau=2>0\\
\Delta=2>0
\end{equation}

\begin{equation}
\tau^2-4\Delta=2^2-4(2)=4-8=-4<0
\end{equation}

 


```python
def fun (t,xy,a,b,c,d):
  x,y=xy
  return [a*x+b*y,c*x+d*y]
  
a,b,c,d=1,-1,1,1

yi=-10; yf=10
xi=-10; xf=10
ti=0; tf=10; dt=1000
tspam=[ti, tf]
t=np.linspace(ti,tf,dt)
for x0 in np.arange(xi,xf,1):
  for y0 in np.arange(yi,yf,1):
    sol=solve_ivp(fun,tspam,[x0,y0],args=(a,b,c,d),t_eval=t)
    plt.plot(sol.y[0], sol.y[1])

plt.axis([-10, 10, -10, 10])
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

# Creating data set
w = 10
Y, X = np.mgrid[-w:w:100j, -w:w:100j]
U = a*X+b*Y
V = c*X+d*Y
speed = np.sqrt(U**2 + V**2)
 
# Creating plot
plt.streamplot(X, Y, U, V, density = 1,color='gray')
 
# show plot
plt.show()
```

##5.3.3 (Fuera de contacto con sus propios sentimientos) 

Supongamos que Romeo y Julieta reaccionan entre sí, pero no consigo mismos: $\dot{R}=aJ , \dot{J}=bR$ ¿Qué pasa?

**Sistema en forma matricial**

\begin{equation}
\left[\begin{matrix}\dot{R}\\\dot{J}\\\end{matrix}\right]=\left[\begin{matrix}0&a\\b&0\\\end{matrix}\right] \left[\begin{matrix}R\\J\\\end{matrix}\right]
\end{equation}

**Polinomio caracteristico**

\begin{equation}
det\left[\begin{matrix}0-\lambda&a\\b&0-\lambda\\\end{matrix}\right] =0
\end{equation}

\begin{equation}
(0-\lambda)(0-\lambda)-(a)(b)=0\\
\lambda^2-ab=0
\end{equation}


**Valores propios**
\begin{equation}
\lambda^2-ab=0\\
\lambda_{1}=\sqrt{ab}\\
\lambda_{2}=-\sqrt{ab}
\end{equation}

**Vectores propios**

\begin{equation}
(A-\lambda_{1}I)\vec{v_1}=\vec{0}\\

##5.3.5 (Guisantes en una vaina) 

Si Romeo y Julieta son clones románticos $\dot{R}=aR+bJ, \dot{J}=bR+aJ$ ¿deberían esperar aburrimiento o felicidad?

**Sistema en forma matricial**

\begin{equation}
\left[\begin{matrix}\dot{R}\\\dot{J}\\\end{matrix}\right]=\left[\begin{matrix}a&b\\b&a\\\end{matrix}\right] \left[\begin{matrix}R\\J\\\end{matrix}\right]
\end{equation}

**Polinomio caracteristico**

\begin{equation}
det\left[\begin{matrix}a-\lambda&b\\b&a-\lambda\\\end{matrix}\right] =0
\end{equation}

\begin{equation}
(a-\lambda)(a-\lambda)-(b)(b)=0\\
(a-\lambda)^2-b^2=0\\
((a-\lambda)-b)((a-\lambda)+b)=0\\
(a-b-\lambda)(a+b-\lambda)=0
\end{equation}


**Valores propios**
\begin{equation}
\lambda_{1}=a+b\\
\lambda_{2}=a-b
\end{equation}

**Vectores propios**

\begin{equation}
(A-\lambda_{1}I)\vec{v_1}=\vec{0}\\
\left[\begin{matrix}a-\lambda_{1}&b\\b&a-\lambda_{1}\\\end{matrix}\right]\vec{v_1}=\vec{0}\\
\left[\begin{matrix}-b&b\\b&-b\\\end{matrix}\right]\vec{v_1}=\vec{0}\\
\vec{v_1}=\left[\begin{matrix}1\\1\\\end{matrix}\right]
\end{equation}


\begin{equation}
(A-\lambda_{2}I)\vec{v_2}=\vec{0}\\
\left[\begin{matrix}a-\lambda_{2}&b\\b&a-\lambda_{2}\\\end{matrix}\right]\vec{v_2}=\vec{0}\\
\left[\begin{matrix}b&b\\b&b\\\end{matrix}\right]\vec{v_2}=\vec{0}\\
\vec{v_2}=\left[\begin{matrix}-1\\1\\\end{matrix}\right]
\end{equation}

**Clasifica el punto fijo en el origen**

\begin{equation}
\tau=2a\\
\Delta=a^2-b^2
\end{equation}

\begin{equation}
\tau^2-4\Delta=(2a)^2-4(a^2-b^2)=4b^2
\end{equation}

si $a<b$ es saddle point\
si $a<b<0$ es stable node\
si $0<b<a$ es unstable node
